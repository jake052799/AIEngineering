```python
# Step D Build a Model while applying Normalization and 3 hidden layers


# Import Packages   
import pandas as pd
import numpy as np
import keras
```

```python
# Load the data
concrete_data = pd.read_csv('concrete_data.csv')
concrete_data
```

```python
# Create X matrix of the predictors and create y matrix as the target
# Also apply normalization to the X/predictor variables

concrete_data_columns = concrete_data.columns

X = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']]
y = concrete_data['Strength']

X_norm = (X - X.mean()) / X.std()

X_norm.head()


from sklearn.model_selection import train_test_split
```

```python
# Number of variables in the X matrix 
n_cols = X_norm.shape[1]
print(n_cols)
```

```python
# Create regression model definition

from keras.models import Sequential
from keras.layers import Dense

def regression_model():
    model = Sequential()
    model.add(Dense(10, activation= 'relu', input_shape= (n_cols,)))
    model.add(Dense(10, activation= 'relu'))
    model.add(Dense(10, activation= 'relu'))
    model.add(Dense(1))

    model.compile(optimizer= 'adam', loss= 'mean_squared_error')
    return model
```

```python
# Use for loop to get a list of 50 Mean Squared Errors with normalization

prediction = []
seed = 2
for i in range(50):
    X_train, X_test, y_train, y_test = train_test_split(X_norm, y, test_size= 0.3, random_state= seed + i * 2)
    model = regression_model()
    model.fit(X_train, y_train, epochs= 50, verbose= 2)
    y_pred = model.predict(X_test)
    from sklearn.metrics import mean_squared_error
    prediction.append(mean_squared_error(y_test, y_pred))
```

```python
prediction
```

```python
# Get average prediction mean mean_squared_error and its standard deivation
prediction_mean = sum(prediction) / len(prediction)
variance = sum([((x - prediction_mean) ** 2) for x in prediction]) / len(prediction)
std = variance ** 0.5

print("The prediciton mean squared error mean is {} \nThe prection standard deivation is {}".format(prediction_mean, std))
```
